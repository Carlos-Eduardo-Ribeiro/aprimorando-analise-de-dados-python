```python
import numpy as np

#Criando dados
nomes = np.array(['João', 'Henrique', 'Jorge', 'Orácio', 'Juliana'])
notasMatematica = np.array([9.66,7.23,4.3,8.33,5.88])
notasFisica = np.array([7.2,3.44,6.55,7.8,9.2])

#concatenendo em colunas e retirando a média 
media = np.mean(np.column_stack((notasMatematica, notasFisica)), axis = 1)
print(media)

#idêntificando alunos aprovados (apro>6)
aprovados = np.where(media>6)
print(aprovados)
print(nomes[aprovados])

#Organizando os dados em uma tabela, o np.column_stack pega os arrays e coloa em colunas(verticalmente)
df = np.column_stack((nomes, notasMatematica, notasFisica, media))
print(df)

#colocando coluna 1 para um pseudo cabeçalho
cabecalho = np.array(['Nome aluno', 'Nota Matemática', 'Nota Física', 'Média'])

df = np.vstack((cabecalho, df))
print(df)

#Alternado posição especifica no array de 2
df[0,0] = 'Nome Aluno'
print(df)
```
