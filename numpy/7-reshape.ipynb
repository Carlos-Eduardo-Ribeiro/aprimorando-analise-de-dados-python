#Reshape
 
- Conversão de uma matriz formado por n(arrays) para um array de ndim diferete

```python
import numpy as np
 
array_1d = np.array([1,3,11,6,88,99,21,45])
print(array_1d.ndim)
print(array_1d.shape)
print(array_1d)
```

```python
#Usando o reshapepara converter um array de uma dimenão (8,0) para uma array de (2,4)
 
#É necesario analisar o tamanho do array(suas diemns~pes linha e coluna) para
#analisar em que configuração aquele array pode estar quando se usar o rashape
 
array_2d = array_1d.reshape(2, 4)
print(array_2d)
```

```python
#Usando o rashape para converter uma matriz de ndim(s) em um array unidimencional
 
array_2d2 =  np.array([[1,2,3], [4,5,6]])
print(array_2d2.ndim)
print(array_2d2.shape)
print(array_2d2)
 
array_1d1 = array_2d2.reshape(6,)
print(array_1d1.ndim)
print(array_1d1.shape)
print(array_1d1)
```

```python
#Simulando uma imagem de 8X8 (64px) cim um array de uma dimensão
 
pixels = np.arange(64)
pixels
 
#usando o reshape para criar uma matriz 2d 8x8
imagem = pixels.reshape(8,8)
print(imagem)
print(imagem.shape)
 
 
#Invertendo os pixels da imaem para simular a inversão de imagens
#Suntarindo 63 de cada posição de um array 8x8
imagem_ivertida = 63 - imagem
print(imagem_ivertida)
```
