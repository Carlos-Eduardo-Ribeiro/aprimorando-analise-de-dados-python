```python
import numpy as np

array_1 = np.array([[1,2,3]])
array_2 = np.array([[4,5,6]])

print(array_1.shape," ", array_2.shape)
print(array_1.ndim, " ", array_2.ndim)
```

```python
#Concatenadndo arrays

nova_array = np.concatenate((array_1, array_2))
nova_array
```

```python
#criando arrays para nomes e notas com Vstack para empilhar verticalemnte(linhas) os dados

nomes = np.array([["Amanda", "Joaquin", "Kaique"]])
notas_Matematica = np.array([[9.6, 9.2, 10]])
notas_Portugues = np.array([[7, 6.5, 9.2]])

resultadoV = np.vstack((nomes, notas_Matematica, notas_Portugues))

print(resultadoV)
```

```python
#Criando array concatendo horizontalmente(colunas) com o np.hstack

nomes = np.array([["Amanda", "Joaquin", "Kaique"]])
notas_Matematica = np.array([[9.6, 9.2, 10]])
notas_Portugues = np.array([[7, 6.5, 9.2]])

resultadoH = np.hstack((nomes, notas_Matematica, notas_Portugues))
print(resultadoH)
```

```python
#Adicionando coluna com m√©dia das notas

media = np.mean(notas_Portugues, axis=1) #axis=1 representa a linha, logo a media vai ser retirada da linha
matrizResultado = np.column_stack((notas_Portugues, media))#O column_stack empilha uma coluna ao lado 
print(matrizResultado)
```
