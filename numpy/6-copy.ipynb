#Copy de Arrays
 
- Quando você usa o copy, uma nova cópia do array é criada na mémoria
 
- O que faz com que as cópiasseja qual for seu ndin seja indepêndetes!
 
#View de Arrays
 
- Não gera uma cópia do array, ele compartilha a versão(dados) do array original.


```python
import numpy as np
 
arrayOriginal = np.array([[2,4,6], [3,11,89]])
print(arrayOriginal.ndim)
print(arrayOriginal.shape)
print(arrayOriginal)
```

```python
#Cópia Rasa (View)
 
array_view = arrayOriginal[0: , :2]
print(array_view)
 
array_view[0, 0] = 100
print(array_view)
 
print(arrayOriginal)
 
#Quando eu uso uma cópia rasa do meu array, atraves de atribuição simples, eu gero um
#espelho do array view para  orginal, se eu alterar o view, altero o original e vise
#versa, o quê não há problema se não existe a necessidade de manter o array original
```

```python
 
#Cópia Profunda
 
array_copia = arrayOriginal.copy()
print(array_copia)
 
array_copia[1, 2] = 100
print(array_copia)
 
print(arrayOriginal)
 
#Por ser uma cópia profunda usando a função (copy), não gera depêndencia entre os arrays ndin
#logo eles são indepeêndentes entre sí, qualquer alteração feita bem um não reflete no outro.
```
