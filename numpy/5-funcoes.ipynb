```python
import numpy as np
```

```python
array = np.array([1,2,3,4,5])
array
```

```python
#Função interação do array sum(soma) para arrays
interacao = sum(array)
interacao
```

```python
#Função soma posição a posição do array
soma = np.add(array, 10)
soma
```

```python
#Função de subtração!
 
subtracao = np.subtract(array, 2)
print(subtracao)
```

```python
#Função de Multiplicação!
 
produto = np.multiply(array, 2)
print(produto)
```

```python
#função de divisão
 
divisao = np.divide(array, 2)
print(divisao)
```

```python
#funções trigonometricas
 
#Seno
seno = np.sin(array)
print(seno)

#Cosseno
cosseno = np.cos(array)
print(cosseno)
```

```python
#Média
media = np.mean(array)
print(media)
 
#Desvio Padrão
desvioPadrao = np.std(array)
print(desvioPadrao)
```

```python
#Funções de comparação ou funções de Mascara
 
comparacaoMaior3 = array > 3
print(comparacaoMaior3)
```

```python
 
#Combinado Funçãoes
 
#Raiz quadrada de cada posição + potênciação de 2 de cada posição
resultado = np.sqrt(array) + np.power(array, 2)
print(resultado)
```

```python
#Funções Expeonenciais
 
#As funções esponeciais usam a a constante de Euler = 2,71828182...
exp = np.exp(array)
print(exp)
```

```python
#Funções de Logaritimo
 
log = np.log(array)
print(log)
```

```python
#Funções de arredondamento
 
arrendondado = np.around(array)
print(arrendondado)
```

```python
#Função de potência
 
pot = np.power(array, 2)
print(pot)
```
