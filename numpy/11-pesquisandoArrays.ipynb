```python
import numpy as np
array = np.array([10,20,30,40,50])
array
```

```python
#Como pesquisar a posição do valor 30

posicao = np.where(array==30)
print(f'{posicao}, {array.dtype}')

#Como achar alunos aprovadios nota>6 em um array de notas

notas = np.array([9.88,7.5,6.4,3.2,6.9])
print(notas)

alunosAprovados = np.where(notas>6) #Quando se usar mascra em um array, ele retorna somente a posição em que as condições da mascar são verdadeiras
print(notas[alunosAprovados]) #Usando atribuição de uma mascara para acessar um valor

#Como achar a nota máxima

notaMaxima = np.max(notas) #Ou np.argmax que retorna o índice do valor máximo 
print(notaMaxima)
```
