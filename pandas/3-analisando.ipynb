```python
import pandas as pd
```

```python
#Criando dados

dadosVendas = {
    'Produto' : ['Calça', 'Camisa', 'Camisa', 'lençou', 'Saia', 'Saia', 'luva', 'Bone', 'Calça', 'Tenis'],
    'Preço' : [50, 20, 40, 90, 100, 30, 20, 10, 20, 80],
    'Quantidade' : [10, 12, 40 ,23, 34, 45, 11, 45, 89, 100],
    'Mês' : ['Jan', 'Fev', 'Fev', 'Abr', 'Jan', 'Jun', 'Abr', 'Jun', 'Nov', 'Jan']
}
```

```python
df = pd.DataFrame(dadosVendas)
df.head(5)
```

```python
#Calculando a receita

df['Receita'] = df['Preço'] * df['Quantidade'] #E interessante que você pode fazer quando gera uma nova coluna fazer de maneria mais formal ou de maneira mais bruta, mas para mascaras sempre lembrar de ter de retornar o df resulatante e não apenas um recorte dele
df.head()
```

```python
#Agrupando receita por produto

resumoVendas = df.groupby('Produto')['Receita'].sum().reset_index()
resumoVendas
```

```python
#Filtra produtos acima de 3000 reais

produtosAcimaTresMil = df[df['Receita']>3000]
produtosAcimaTresMil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize =(8,5))
plt.bar(resumoVendas['Produto'], resumoVendas['Receita'], color='skyblue')
plt.title('Vendas')
plt.xlabel('Produtos')
plt.ylabel('Valor Vendido')
```
