```python
#1-Criando uma series com dados de vendas mensais

vendas = pd.Series([100, 22, 45, 66, 66, 22, 45, 62, 78, 90, 23, 10], index = ['jan', 'Fev', 'Mar', 'Abr', 'Mai', 'jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
print(vendas) #E necessario passar um idex para cada valor da serie se o index for declarado
```

```python
#2-Calcular e exibir a media de vendas

mediaVendas = vendas.mean()
print(mediaVendas)
```

```python
#3-Mês com  maior venda

maiorMes = vendas.idxmax() #Retornar o index com a posição de memoria na series que possue o maior valor
print(maiorMes)
```

```python
#4-Exibir as vendas acima da media

vendasAcimaMedia = vendas[vendas>mediaVendas] #A condição(mascara) deve estar dentro da estrutura de memoria pois ela intera considerndo as posições e a estrutara em questão(series)
print(vendasAcimaMedia)
```
