```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importando base de dados em csv
dados = pd.read_csv(r"..\dados\produtos.csv") #Os dois
dados
```

```python
dados.head(10) #dez primeiros produtos
```

```python
dados.tail(10) #dez ultimos produtos
```

```python
#Realisando análise rápida acerca de qual categoria vende mais.

qtdVendasPorProduto = dados.groupby('Categoria')['Quantidade'].sum().reset_index()
qtdVendasPorProduto
qtdVendasPorProduto.plot(kind='bar', x='Categoria', y='Quantidade', title='Total Vendido', xlabel='Categorias', ylabel='Quantidade', color='blue', rot=0)
```
