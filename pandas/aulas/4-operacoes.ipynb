```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Nome' : ['Alice', 'João', 'Charlie', 'Emily', 'Julio'],
    'Idade' : [22, 45, 55, 18, 25],
    'Cargo' : ['Analista', 'Coordendor', 'Diretor', 'Desenvolvendor', 'Desenvolvendor'],
    'Salario' : [5500, 6800, 10200, 2500, 3200]
}
```

```python
df = pd.DataFrame(data)
df.head(2) #retornando primeiras 2 linhas
```

```python
df.tail(2)#Pegando as ultimas 2 linhas
```

```python
df.info() #Informações gerais acerca do dataframe
```

```python
df.describe()#Retorna algumas medidas estatísticas acerca do df
```

```python
dfMC = df[df['Salario']>5000] #A mascara esta sendo passndo dentro do proprio df, logo está retornado um df resultant ee nãoo um recorte
dfMC
```

```python
dfOrdenadoIdade = df.sort_values(by='Idade', ascending=True) #Ordenando pela idade.É interessante a questão de realização de chamadas de funções em comparação com operações diretamente no df seja para gera uma coluna ou fazer alguma modificação
dfOrdenadoIdade
```

```python
recorteIdade = df[['Nome', 'Idade']] #fazendo recorte de um df para um df resultante com colunas objetivo
recorteIdade
```

```python
df['Bonus'] = df['Salario'] * 0.1 #criando uma coluna com base em 10% da coluna salario
df['Bonus'] = df['Bonus'] + df['Salario'] #Mesclando colunas
df = df.rename(columns = {'Bonus': 'Salario+Bonus'}) #renomeando uma coluna
df
```

```python
dfAgrupado = df.groupby('Cargo')['Salario'].mean().reset_index() #Agrupando salarios por cargo e retirando a média
dfAgrupado
```

```python
plt.Figure(figsize=(9, 5))
plt.bar(df['Cargo'], df['Salario'], color='red')
```

```python
#Usando o plot do próprio df para plotar um gráfico barra. Obs:É necessário importar o matplotlib pois ele é usando pelo plot 
df.plot(kind='bar', x='Cargo', y='Salario', color='red', title = 'Salário por Cargo', xlabel='Valor Salário', ylabel='Cargo', rot=45, figsize=(9,8))
```
