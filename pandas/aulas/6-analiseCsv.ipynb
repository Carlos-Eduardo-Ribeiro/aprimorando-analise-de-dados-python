```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfprodutos = pd.read_csv(r"..\dados\produtos.csv")
dfprodutos
```

```python
dfprodutos.describe() #Descrição estatística
dfprodutos.info() #Descrição Geral
dfprodutos.shape #Corpo do df (linhas, colunas)
dfprodutos.ndim #Número de dimensões
dfprodutos.dtypes #Tipos de cada coluna 
```

```python
dfprodutos['Quantidade'] = pd.to_numeric(dfprodutos['Quantidade'])  #convertendo a coluna quantidade para uma do tipo numeric 
dfprodutos['Preço'] =  pd.to_numeric(dfprodutos['Preço']) #convertendo a coluna preço para uma do tipo numeric
dfprodutos.dtypes
```

```python
#1-Desempenho de vandas por categorias

#Sempre crie um sub df para preservar o original

vendascategorigas =  dfprodutos.groupby('Categoria')['Quantidade'].sum()
vendascategorigas
```

```python
#2-produto mais Vendido

produtoMaisVendido = dfprodutos.groupby('Produto')['Quantidade'].sum().idxmax()#Ele considera o idmax como o registro pelo qual ele e agrupado
produtoMaisVendido
```

```python
#3-Preço médio de cada produto

mediaPrecoProduto = dfprodutos.groupby('Produto')['Preço'].mean().reset_index()
mediaPrecoProduto
```

```python
#4-Corrlação entre qtd de unidades e preço unitário

correlacao =  dfprodutos['Quantidade'].corr(dfprodutos['Preço']) #Avaliando correlação preço/quantidade
correlacao
```

```python
#5-Exportando desempenho de vendas por categorias para um csv

desVendasCategorigas = vendascategorigas.reset_index()
desVendasCategorigas.to_csv("../dados/vendas_categoria.csv", index=False)

```
