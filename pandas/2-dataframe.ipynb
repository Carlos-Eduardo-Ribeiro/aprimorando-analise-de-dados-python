```python
dadosNotas = {
    'Nome' : ['Tango', 'Maria', 'Julia'],
    'Matematica' : [6.7, 8.0, 9.0],
    'Geografia' : [9.2, 8.3, 4.2]
}
```

```python
dfNotas = pd.DataFrame(dadosNotas)
dfNotas.head()
```

```python
#Criando uma nova coluna no df com a media das notas por aluno

dfNotas['Media'] = dfNotas[['Matematica', 'Geografia']].mean(axis=1)#Quando eu não especifico uma coluna existente no df ele me gera uma nova
dfNotas.head()
```

```python
#Identificando alunos aprovados com nota maior que 8

alunoAprovados  = dfNotas[dfNotas['Media']>8] #E necessario especificar a macara com o proprio df e coluna dentro do df pois assim ele retornara um df resultante e não um recorte do df que não e um df
alunoAprovados.head()
```

```python
#Exibir a media e os alunos aprovados

dfAprovados = dfNotas[['Nome', 'Media']]
dfAprovados.head()
```
