```python
#Questão7

vogais = 'aeiou'
cont = 0
palavra = str(input('Digite a palavra: '))
palavra = palavra.lower()
for i in palavra:
    if i in vogais:
        cont+=1
print(f'Existe {cont} vogais')
```

```python
#Questão8

import statistics

resposta = 1
listaNotas = []
while resposta != 0:

    nota = float(input('Digite a sua nota: '))

    if nota >= 0 and nota<=10:
       listaNotas.append(nota)
    else:
       print('\nVocê digitou um valor fora do range (0 a 10)!')

    valor = int(input(f'\nGostaria de continuar ?\nSim = 1\nNão = 0'))
    if valor == 1:
        resposta = 1
    else:
        resposta = 0
try:
  print(f'A média é {statistics.mean(listaNotas)}')
except:
   print(f'\nVocê não digitou nenhum valor!')
```
