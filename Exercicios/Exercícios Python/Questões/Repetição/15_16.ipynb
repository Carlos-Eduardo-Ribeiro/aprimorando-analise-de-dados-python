```python
#Questão15

n = int(input('Digite o valor: '))

for i in range(n):
    print('')
    for j in range(n): 
        if (i+j)%2==0:
            print('#', end=' ')
        elif (i+j)%2!=0:
            print('.', end=' ')
```
