```python
#Quest達o11

valor = int (input('Digite um valor inteiro para o fatorial: '))
valorBase = valor
for i in range(valor-1):
    valorBase *= (valor - (i+1))

print(f'Resultado do fatorial de {valor} e igual a {valorBase}') 
```

```python
#Quest達o12

valor = int(input('Digite o valor: '))
cont = 1
for i in range (valor):
    valor = valor//10
    if valor > 0:
        cont+=1
    elif valor == 0:
        continue
    
print(f'Resultado: {cont}')
```

```python
#Quest達o13

lista = []
valor = ""

while valor !=0:
    valor = float(input('\nDigite o valor: '))
    lista.append(valor)

print(f'O maior valor {max(lista)}, o menor valor {min(lista)}')
```
