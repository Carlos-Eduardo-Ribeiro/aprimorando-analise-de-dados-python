```python
#Questão17

numero = int(input('Digite o valor para o teste: '))
cont = 0

for i in range(numero): 
    if numero%(i+1)==0:
        cont+=1
        
if cont > 2:     
    print(f'O numero {numero} não primo')
else:
    print(f'O numero {numero} não e primo')
```

```python
#Questão18

pares = []
impares = []

while True:

    valor = int(input('Digite um valor: '))

    if valor==0:
        break
    if valor%2==0:
        pares.append(valor)
    else:
        impares.append(valor)

print(f'Soma valores pares: {sum(pares)}\nSoma valores impares: {sum(impares)}')
```

```python
#Questão19

n = int(input('Digite um valor: ')) 

for i in range(n):
    print('')
    for j in range(n):
        if i == j:
            print(1, end =' ')
        else:
            print(0, end=' ')
```
