```python
#Questão31

def avaliacaoLogical(x,y,z):

    tz = z.lower()
    tz = tz.strip()
    operadorespermitidos =['or', 'and', 'xor']
    zeroUm = [1,0]

    if tz not in  operadorespermitidos:
            return f'O operador {z} não existe!'
    elif x not in zeroUm or y not in zeroUm:
            return f'O os valores {x} ou {y} não existe!'

    match tz:
        case 'or':
            if x==1 or y==1:
                return 'True'
            else:
                return 'False'
        case 'and':
            if x==0 or y==0:
                return 'False'
            else:
                return 'True'
        case 'xor':
            if x==0 and y==0 or x==1 and y==1:
                return 'True'
            else:
                return 'False'
        case '' :
            return 'Nenhum valor imputado'

valorX = int(input('Digite o valor de x: '))
valorY = int(input('Digite o valor de y; '))
operador = str(input('\nDigite o operador desejado (and, or, xor:): '))  
print(f'\nO resultdo de {valorX} {operador} {valorY} é {avaliacaoLogical(valorX,valorY,operador)}') 
```
