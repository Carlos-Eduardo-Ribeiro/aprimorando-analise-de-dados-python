```python
#Questão01

def is_even(n):
    if n%2==0:
        return True
    else:
        return False

valor = int(input(f'Digite um valor inteiro: '))
print(f'Resultado {is_even(valor)}')
```

```python
#Questão02

import math

def area_circle(r):
    return round(math.pi * math.pow(r, 2), 2)

valor = int(input(f'Digite um valor inteiro: '))
print(f'Resultado {area_circle(valor)}')
```

```python
#Questão03

letras = 'aeiouAEIOU'

resul = lambda x:sum(x.count(l) for l in letras)

palavra = str(input(f'Digite uma palavra: '))
print(f'Resultado {resul(palavra)}')
```

```python
#Questão04

def clamp(x, lo, hi):
    if x<lo:
        return lo
    elif x>hi:
        return hi
    else:
        return x
    
valorX = int(input(f'Digite um valor inteiro para x: '))
valorLo = int(input(f'Digite um valor inteiro para lo: '))
valorHi = int(input(f'Digite um valor inteiro para hi: '))
print(f'Resultado {clamp(valorX, valorLo, valorHi)}')
```

```python
#Questão05

def toMinuscula(l): 

    primeiraLetra = l[0]
    
    if l.isalpha():
        return primeiraLetra.lower()
    else:
        return 'None'

palavra = str(input(f'Digite uma palavra: '))
print(f'Resultado "{toMinuscula(palavra)}"')
```
