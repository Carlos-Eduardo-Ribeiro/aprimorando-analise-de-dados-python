```python
#Questão11

import string

def validadorSenha(s):

    alfMai = string.ascii_uppercase
    alfMin = string.ascii_lowercase
    x = y = False

    if len(s)>8:
        for i in s:
         if(i in alfMai):
            x = True
         elif(i in alfMin):
            y = True
         elif(x == True and y == True):
            continue
         
    if (x == True and y == True):
       return True
    else:
       return False
    
senha = str(input('Digite a sua senha: '))
print(f'Resultado: {validadorSenha(senha)}')
```

```python
#Questão12

import statistics as st

def mode(lista):
    return st.mode(lista)

lista = []
qtd = int(input('Digite a quantidade de números que deseja: '))
for i in range (qtd):
    valor = int(input(f'\nDigite o valor {i + 1}'))
    lista.append(valor)
print(f'\nO resultado é {mode(lista)}')
```
