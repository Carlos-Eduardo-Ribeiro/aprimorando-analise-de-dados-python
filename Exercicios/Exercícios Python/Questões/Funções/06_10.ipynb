```python
#Questão06

import datetime

def formater(s):
    min, sec = divmod(s, 60)
    hora, min = divmod(min, 60)
    print(f'{hora:02d}:{min:02d}:{sec:02d}')

valor = int(input('Digite o valor: '))
formater(valor)
```

```python
#Questão07

def grade(score):
    if score<0 or score>10:
        return 'Nota invalida'
    elif score < 3:
        return 'F'
    elif score < 5:
        return 'D'
    elif score < 7:
        return 'C'
    elif score < 9:
        return 'B'
    else: 
        return 'A'
    
valor = int(input('Digite o valor: '))
print(f'Resultado: {grade(valor)}')
```

```python
#Questão08

def divisaoSegura(x, y):
    if y==0:
        return 'Divisão por zero não permitida'
    else:
        return x/y
    
valor1 = int(input('Digite o valor 1: '))
valor2 = int(input('Digite o valor 2: '))
print(f'Resultado: {divisaoSegura(valor1, valor2)}')
```

```python
#Questão09

def somaList(l):
    if(len(l)==0):
        return 'Lista vazia'
    else:
        soma = 0
        for i in l:
          soma += i 
        return soma
    
lista = [1,2,3,4,5]
print(somaList(lista))
```
